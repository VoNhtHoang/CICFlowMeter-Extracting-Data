```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return self.name

class Dog(Animal):
    def bark(self):
        return "Woof!"

class Cat(Animal):
    def meow(self):
        return "Meow!"

# Sử dụng lớp con
dog = Dog("Buddy")
print(dog.speak())  # Kết quả: Buddy
print(dog.bark())   # Kết quả: Woof!

cat = Cat("Whiskers")
print(cat.speak())  # Kết quả: Whiskers
print(cat.meow())   # Kết quả: Meow!
```

```python


dict_df = {"Benign": 10, "DDoS" : 30, "DoS": 20, "Reconn": 5, "MITM": 50}
dict_predf = {"Benign": 300, "DDoS" : 300, "DoS": 300, "Reconn": 60}

dict_newdf = {}
for key, val in dict_df.items():
    if key not in dict_predf.keys():
        dict_newdf[key]=val
    else:
        dict_newdf[key]= dict_predf[key]
        
num_labels = len(dict_newdf)
print(num_labels)
target_per_label = 1200 // num_labels
print(target_per_label)

delta_samples = 0
count_delta = 0
for key, val in dict_predf.items():
    if val < target_per_label:
        delta_samples += target_per_label - val
        count_delta +=1

print(delta_samples // (len(dict_predf)-count_delta))
for key, val in dict_predf.items():
    if val >= target_per_label:
        print(True, key)
        dict_newdf[key] = target_per_label + delta_samples // (len(dict_predf)-count_delta)
    else:
        dict_newdf[key] = val
        
print(dict_newdf)
```
